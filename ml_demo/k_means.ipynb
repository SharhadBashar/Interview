```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Step 1: Create synthetic data
# Features: Square footage and Price (in $1000s)
data = np.array([
    [1200, 300],
    [1500, 350],
    [900, 200],
    [2000, 500],
    [800, 150],
    [2500, 600],
    [1000, 220],
    [2200, 550],
    [1700, 400],
    [1300, 320]
])
```

```python
# Step 2: Apply K-Means Clustering
# Define the number of clusters
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(data)

# Get cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Step 3: Visualize the clusters
plt.figure(figsize=(8, 6))
colors = ['red', 'blue', 'green']
for i in range(num_clusters):
    cluster_points = data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {i+1}", color=colors[i])
    
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=300, c='yellow', marker='X', label='Centroids', edgecolor='black')
plt.title("K-Means Clustering: Properties Grouping")
plt.xlabel("Square Footage")
plt.ylabel("Price ($1000s)")
plt.legend()
plt.show()
```

```python
# Step 1: Create synthetic data
data = np.array([
    [1, 2], [1, 3], [2, 2], [2, 3], [3, 3], [3, 2],   # Cluster 1
    [6, 5], [6, 6], [7, 5], [7, 6],                   # Cluster 2
    [8, 1], [8, 2], [9, 1], [9, 2], [10, 2]           # Cluster 3
])
```

```python
# Step 2: Visualize data before clustering
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], color='black')
plt.title("Before Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
```

```python
# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Step 4: Visualize data after clustering
plt.subplot(1, 2, 2)
colors = ['red', 'blue', 'green']
for i in range(3):
    cluster_points = data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[i], label=f"Cluster {i+1}", alpha=0.6)
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='X', s=100, label='Centroids')
plt.title("After Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

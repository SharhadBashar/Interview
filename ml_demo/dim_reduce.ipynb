```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
# Step 1: Load the Iris dataset
data = load_iris()
X = data.data  # Features: Sepal length, Sepal width, Petal length, Petal width
y = data.target  # Target labels: Species
```

```python
# Step 2: Apply PCA to reduce dimensions
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X)
```

```python
# Step 3: Visualize the reduced data
plt.figure(figsize=(8, 6))
for label, color in zip(np.unique(y), ['red', 'blue', 'green']):
    plt.scatter(X_pca[y == label, 0], X_pca[y == label, 1], label=data.target_names[label], color=color, alpha=0.7)
    
    plt.title("PCA Dimensionality Reduction")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()
```
